```python
#pip install mysql-connector-python

import pandas as pd
import mysql.connector
```

# Load Data from a MySQL Database into a Pandas DF

```python
#mysql.connector is the official connector for Python from MySQL

#setting up the engine
sql_engine = mysql.connector.connect(user='root', password='password', host='localhost',
                                     port = 3306, database='supermarket')

mycursor = sql_engine.cursor()

#Here's the query
sql =   """
           SELECT gender, `Customer type`, COUNT(1) Count
            FROM supermarket_sales
            GROUP BY gender, `Customer type`
        """

mycursor.execute(sql)

#Loading rows from Database into Pandas DataFrame
table_rows = mycursor.fetchall()
df = pd.DataFrame(table_rows)

#Commit changes and close connection
sql_engine.commit()
sql_engine.close()

df.head()
```

# DDL & DML Commands on MySQL using Python (Data Manipulation)

```python
#setting up engine
sql_engine = mysql.connector.connect(user='root', password='password', host='localhost', 
                                     port = 3306, database='supermarket')

mycursor = sql_engine.cursor()

#The "INSERT" command below can be any DDL or DML command you want
#just make sure it's in quotation marks, 3 quotes if it's multiple lines
sql = "INSERT INTO sms_new_table (`Invoice ID`, `Date`)  VALUES ('03-3333', '01/01/2021')"

mycursor.execute(sql)

#Commit changes and close connection
sql_engine.commit()
sql_engine.close()

print(mycursor.rowcount, "record inserted.")
```

# The SQL Alchemy method

```python
#pip install sqlalchemy
#pip install pymysql

import sqlalchemy as sa

#connection engine
mysql_engine = sa.create_engine('mysql+pymysql://root:password@localhost:3306/supermarket')

#loading data into Pandas DataFrame
sms = pd.read_sql_table('supermarket_sales', mysql_engine)

sms.head()
```

```python
#Loading a Pandas DataFrame into MySQL 
#the 'sms_new_table' below is the name of the table being inserted into MySQL where 'sms' is the DataFrame

sms.to_sql('sms_new_table', mysql_engine)
```
